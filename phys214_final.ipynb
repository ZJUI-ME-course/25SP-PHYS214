```python
# 
import matplotlib.pyplot as plt

# 定义能带参数
lower_band = {"energy": 0, "color": "blue", "label": "Lower band"}
isolated_band = {"energy": 2, "width": 1, "color": "red", "label": "Isolated band"}
higher_band = {"energy": 4, "color": "green", "label": "Higher band"}

# 绘制能带（水平线）
k = [0, 1]  # 波矢示意范围
for band in [lower_band, isolated_band, higher_band]:
    center = band["energy"]
    if band == isolated_band:
        # 孤立能带显示带宽（上下延展）
        plt.fill_between(k, center - band["width"], center + band["width"], 
                        color=band["color"], alpha=0.3, label=band["label"])
    else:
        plt.plot(k, [center] * 2, color=band["color"], linewidth=2, label=band["label"])

# 绘制能隙（黑色虚线）
gap_positions = [(lower_band["energy"], isolated_band["energy"] - isolated_band["width"]), 
                 (isolated_band["energy"] + isolated_band["width"], higher_band["energy"])]
for y_bottom, y_top in gap_positions:
    plt.plot([0.5, 0.5], [y_bottom, y_top], 'k--', linewidth=1)

# 标注费米能级
initial_fermi = 1
final_fermi = isolated_band["energy"] + isolated_band["width"]
plt.hlines(initial_fermi, *k, colors='purple', linestyles='dashed', label='Initial $E_F$')
plt.hlines(final_fermi, *k, colors='cyan', linestyles='dashed', label='Final $E_F$')

# 美化图表
plt.xlim(-0.2, 1.2)  # 扩展波矢范围以显示费米线
plt.ylim(-1, 6)       # 扩展能量范围
plt.xlabel("Wave vector (k)", fontsize=12)
plt.ylabel("Energy (E)", fontsize=12)
plt.title("Band Filling Process with Isolated Band", fontsize=14)
plt.legend(loc='upper right', framealpha=0.9)
plt.grid(alpha=0.2)
plt.show()
```
